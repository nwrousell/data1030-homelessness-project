## Model Evaluation

In this notebook, you can find the code for:
1. Plotting model scores and compare to baseline
2. Evaluating feature importance (Globally)

#### Read dataset and model results and import required packages

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import pickle
from sklearn.model_selection import ParameterGrid, TimeSeriesSplit, train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
import shap
pd.options.mode.chained_assignment = None
sns.set()
shap.initjs()

df = pd.read_csv("../data/dataset_ftr_eng.csv")

with open('../results/xgboost_results.pickle', 'rb') as f:
    xgboost_results = pickle.load(f)

with open('../results/limited_dataset_results.pickle', 'rb') as f:
    limited_dataset_results = pickle.load(f)
```

#### Setup relevant functions

```python
def clean_up_dataset(df):
    columns_to_drop = ["econ_urb_urbanicity", # redundant information already contained in 'major_city', 'suburban', and 'rural'
                       "urban_cat",           # same as above
                       "urbanicity",          # ditto
                       "coctag",              # id that doesn't carry information
                       "panelvar",            # same as above
                       "state"]               # same information as 'state_abr'
    df = df.drop(columns_to_drop, axis=1)
    
    target_variable="all_per_10k"
    to_remove = df[df[target_variable].isnull()].index
    # print(df.loc[to_remove])
    df = df.drop(index=to_remove)
    
    for item in to_remove:
        # print("checking item: ", item)
        assert(item not in df.index)
    
    perc_missing_per_ftr = df.isnull().sum(axis=0)/df.shape[0]
    fully_missing_ftrs = perc_missing_per_ftr[perc_missing_per_ftr == 1].index
    df = df.drop(columns=fully_missing_ftrs)
    return df

def create_time_lagged_features(df, target_variable="all_per_10k", n_lags=3):
    """
        Returns df with time-lagged features going back `n_lags` years. 
        
        Removes rows without complete data, so lower `n_lags` values will result in more rows.

        Returns X, y
    """
    CoCs = df["cocnumber"].unique()
    Xs = []
    ys = []
    for CoC in CoCs:
        coc_df = df[df["cocnumber"] == CoC]
        coc_y = coc_df[target_variable]
        lags = []
        for i in range(1, n_lags + 1):
            lag = coc_df.shift(i)
            lag.columns = "t-" + str(i) + "_" + lag.columns
            lags.append(lag)
        coc_X = pd.concat(lags, axis=1)
        Xs.append(coc_X.iloc[n_lags:]) # only predict for the years when we have data going far enough back
        ys.append(coc_y.iloc[n_lags:])
        # print("finished "+CoC)
    
    X = pd.concat(Xs, axis=0)
    y = pd.concat(ys, axis=0)

    # because of the removal of some rows, some features may be missing all of their values
    # these features need to be removed
    mask = X.isnull().sum()/X.shape[0] == 1
    missing_features = X.columns[mask]
    X = X.drop(columns=missing_features)

    # remove rows when we don't have the target variable
    idx = y[y.isnull() == True].index
    y = y.drop(index=idx)
    X = X.drop(index=idx)
    
    ## TESTING
    assert(X.shape[0] == y.shape[0]) # Shapes match up ✓

    cocs_to_check = [f"t-{i}_cocnumber" for i in range(1,n_lags+1)]
    cocnumbers = X[cocs_to_check]
    assert(cocnumbers.eq(cocnumbers.iloc[:, 0], axis=0).all().all() == True) # time-lagged cocnumbers match ✓

    # correct_order = ((X["t-1_year"] > X["t-2_year"]) & (X["t-2_year"] > X["t-3_year"])).all()
    # assert(correct_order == True) # time-lagged years are in correct order ✓
    
    return X, y

def perform_split(X, y, n_lags=3, years_in_data=8):
    starting_year, ending_year = 2010, 2010 + (years_in_data - 1)
    n_available_years = years_in_data - n_lags
    available_years = list(range(starting_year + n_lags, ending_year + 1))
    assert(len(available_years) == n_available_years)
    
    # calculate which years to put in each set
    n_train_years = np.floor(n_available_years * 0.6).astype(int)
    n_val_years = np.floor((n_available_years - n_train_years) * 0.5).astype(int)
    n_test_years = n_available_years - n_train_years - n_val_years
    # print(n_train_years,n_val_years,n_test_years, available_years)
    
    train_years = available_years[:n_train_years]
    val_years = available_years[n_train_years:(n_train_years+n_val_years)]
    test_years = available_years[(n_train_years + n_val_years):]
    # print(train_years, val_years, test_years)
    
    assert(n_train_years + n_val_years + n_test_years == n_available_years)
    assert(n_val_years > 0)
    assert(n_test_years > 0)
    
    # shift each year by 1, as we split by t-1_year
    train_years_m1 = [v-1 for v in train_years]
    val_years_m1 = [v-1 for v in val_years]
    test_years_m1 = [v-1 for v in test_years]
        
    # perform the split
    mask = X["t-1_year"].isin(train_years_m1)
    X_train = X[mask]
    y_train = y[mask]
    
    mask = X["t-1_year"].isin(val_years_m1)
    X_val = X[mask]
    y_val = y[mask]
    
    mask = X["t-1_year"].isin(test_years_m1)
    X_test = X[mask]
    y_test = y[mask]

    # because of the splitting of some rows,some features may be missing all of their values in the training set
    # these features need to be removed from all sets
    mask_train = (X_train.isnull().sum()/X_train.shape[0] == 1)
    # mask_val = (X_val.isnull().sum()/X_val.shape[0] == 1)
    # mask_test = (X_test.isnull().sum()/X_test.shape[0] == 1)
    # combined_mask = ((mask_train | mask_val) | mask_test)
    missing_features = X_train.columns[mask_train]
    X_train = X_train.drop(columns=missing_features)
    X_val = X_val.drop(columns=missing_features)
    X_test = X_test.drop(columns=missing_features)

    return X_train, y_train, X_val, y_val, X_test, y_test

def make_preprocessor(n_lags, columns):
    onehot_ftrs = ['cocnumber', 'state_abr', 'census_division', 'census_region']
    
    def get_timelag_feature_names(list, n_lags):
        ftrs = []
        for i in range(1, n_lags+1):
            ftrs = ftrs + [f't-{i}_' + s for s in list]
        return ftrs

    
    onehot_ftrs = get_timelag_feature_names(onehot_ftrs, n_lags)
    # std_ftrs = get_timelag_feature_names(std_ftrs, n_lags)
    std_ftrs = [ele for ele in columns if ele not in onehot_ftrs]
    assert([value for value in onehot_ftrs if value in std_ftrs] == []) # assert no overlap
    
    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant',fill_value='missing')),
        ('onehot', OneHotEncoder(sparse_output=False,handle_unknown='ignore')),
        ('std', StandardScaler())])
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('std', StandardScaler(), std_ftrs),
            ('onehot', categorical_transformer, onehot_ftrs)
        ])

    return preprocessor

def setup_data(df, n_lags, years_in_data, target_variable, as_df=False):
    X, y = create_time_lagged_features(df, target_variable=target_variable, n_lags=n_lags)
    

    # perform split
    X_train, y_train, X_val, y_val, X_test, y_test = perform_split(X, y, n_lags=n_lags, years_in_data=8)
    
    # create preprocessor
    preprocessor = make_preprocessor(n_lags, columns=X_train.columns)

    # preprocess data
    X_train_prep = preprocessor.fit_transform(X_train)
    X_val_prep = preprocessor.transform(X_val)
    X_test_prep = preprocessor.transform(X_test)

    if as_df:
        X_train_prep = pd.DataFrame(X_train_prep, columns=preprocessor.get_feature_names_out())
        X_val_prep = pd.DataFrame(X_val_prep, columns=preprocessor.get_feature_names_out())
        X_test_prep = pd.DataFrame(X_test_prep, columns=preprocessor.get_feature_names_out())
        y_train = pd.DataFrame(y_train, columns=[target_variable]).reset_index(drop=True)
        y_val = pd.DataFrame(y_val, columns=[target_variable]).reset_index(drop=True)
        y_test = pd.DataFrame(y_test, columns=[target_variable]).reset_index(drop=True)
    
    return X_train_prep, y_train, X_val_prep, y_val, X_test_prep, y_test
```

## 1. Plot model scores

```python
n_lags = xgboost_results["n_lags"]
models = xgboost_results["models"]
best_params = xgboost_results["best_params"]
xgb_scores = xgboost_results["test_scores"]

clean_df = clean_up_dataset(df)
shap_values = []
for n in n_lags:
    X_train_prep, y_train, X_val_prep, y_val, X_test_prep, y_test = setup_data(df, n_lags=n, years_in_data=8, target_variable="all_per_10k", as_df=True)
    
    # compute baseline 1: mean
    baseline_1_pred = np.full(y_test.shape[0], y_train.mean())
    baseline_1_rmse = mean_squared_error(y_test, np.full(y_test.shape[0], y_train.mean()), squared=False)

    # compute baseline 2: last year's rate
    X, y = create_time_lagged_features(df, target_variable="all_per_10k", n_lags=n)
    X_train, y_train, X_val, y_val, X_test, y_test = perform_split(X, y, n_lags=n, years_in_data=8)

    baseline_2_pred = X_test["t-1_all_per_10k"]
    baseline_2_pred.loc[baseline_2_pred.isnull()] = y_train.mean()
    baseline_2_rmse = mean_squared_error(y_test, baseline_2_pred, squared=False)

    # xgb_mean = np.mean(xgb_scores[n-1])
    test_scores = pd.DataFrame({
        "XGBoost": xgb_scores[n-1],
        "lasso": limited_dataset_results[n-1]["lasso"]["test_score"],
        "ridge": limited_dataset_results[n-1]["ridge"]["test_score"],
        "elastic_net": limited_dataset_results[n-1]["elastic_net"]["test_score"],
        "random_forest": limited_dataset_results[n-1]["random_forest"]["test_score"],
        "KNN": limited_dataset_results[n-1]["KNN"]["test_score"],
        "SVR": limited_dataset_results[n-1]["SVR"]["test_score"],
    })
    # print(test_scores)
                               
    ax = sns.barplot(test_scores, errorbar="sd")
    ax.axhline(y = baseline_1_rmse, label="Baseline Score (mean)", ls="dashed", color="red")
    ax.axhline(y = baseline_2_rmse, label="Baseline Score (predict last year)", ls="dashed", color="blue")
    plt.legend()
    plt.xlabel("Models")
    plt.ylabel("Test Score (RMSE)")
    plt.xticks(rotation=20, ha='right')
    plt.title(f"Test scores with n_lags={n}")
    plt.savefig(f"../figures/test_scores_n_lag_{n}.png", dpi=500, bbox_inches='tight')
    plt.show()
```

## 2. Global Feature Importances

### XGBoost Feature Metrics

```python
metrics = ["weight", "gain", "cover", "total_gain", "total_cover"]

for n in n_lags:
    X_train_prep, y_train, X_val_prep, y_val, X_test_prep, y_test = setup_data(df, n_lags=n, years_in_data=8, target_variable="all_per_10k", as_df=True)
    m = models[n-1][0]
    for metric in metrics:
        scores = m.get_booster().get_score(importance_type=metric)
        scores = pd.Series(scores)
        top_10 = scores.sort_values()[::-1].head(10)
        sns.barplot(x=top_10.values, y=top_10.index)
        plt.title(f"XGBoost Feature Importance: {metric} (n_lags: {n})")
        plt.ylabel("Feature")
        plt.xlabel("Score")
        plt.savefig(f"../figures/xgb_ftr_importance_{metric}_n_lags_{n}.png", dpi=500, bbox_inches='tight')
        plt.show()
```
